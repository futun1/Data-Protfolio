```python
import matplotlib.pyplot as plt

dataset['Attrition'].value_counts().plot(kind='barh' , color=['steelblue','salmon'])
plt.title('Employee Attrition distribution')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
dataset['Age'].plot(kind='hist',bins=20,color='steelblue',edgecolor='white')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('counts')
plt.tight_layout()
plt.show()
```

```python
dataset.groupby('Department')['MonthlyIncome'].mean().plot(kind='bar',color=['steelblue' ,'salmon','mediumseagreen'],edgecolor='white')
plt.title('Avarege Montholy Income by Department')
plt.xlabel('Department')
plt.ylabel('Monthly Income')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
from matplotlib import colors
left=dataset[dataset['Attrition']=='Yes']
left['BusinessTravel'].value_counts().plot(kind='pie' ,autopct='%1.1f%%', colors=['steelblue','salmon','mediumseagreen'])
plt.title('Business Travel Among Employee who left')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
